```python
import requests, pandas as pd
from io import StringIO
import yfinance as yf
```

```python
api = 'https://en.wikipedia.org/w/api.php'
indice = 'IPC'
if indice == 'SP500':
  params = {
      'action': 'parse',
      'page': 'List_of_S&P_500_companies',
      'prop': 'text',
      'format': 'json'
  }
  headers = {'User-Agent': 'indice-tickers-tutorial/1.0'}
  r = requests.get(api, params=params, headers=headers)
  html = r.json()['parse']['text']['*']

  df = pd.read_html(StringIO(html))[0]
  tickers = df['Symbol'].astype(str).tolist()
elif indice == 'IPC':
  params = {
      'action': 'parse',
      'page': 'Índice_de_Precios_y_Cotizaciones',
 #     'section': '1',
      'prop': 'text',
      'format': 'json'
  }
  tickers = ['AC.MX', 'ALSEA.MX', 'AMXB.MX', 'ASURB.MX', 'BBAJIOO.MX', 'BIMBOA.MX', 'BOLSAA.MX', 'CEMEXCPO.MX', 'CHDRAUIB.MX', 'CUERVO.MX', 'GAPB.MX', 'GCARSOA1.MX', 'GCC.MX', 'GENTERA.MX', 'GMEXICOB.MX', 'GRUMAB.MX', 'KIMBERA.MX', 'LABB.MX', 'LACOMERUBC.MX', 'LIVEPOLC1.MX', 'MEGACPO.MX', 'OMAB.MX', 'ORBIA.MX', 'PINFRA.MX', 'Q.MX', 'RA.MX', 'TLEVISACPO.MX', 'VESTA.MX']
```

```python
print(tickers)
```

```python
fecha_inicio = "2026-02-20"
fecha_fin = "2026-02-28"

df_indices = {}

for tick in tickers:
  try:
    info = yf.Ticker(tick)
    df = info.history(start=fecha_inicio, end=fecha_fin)
    df.index = df.index.strftime('%Y-%m-%d')
    df_indices[tick] = df["Close"]

  except Exception as e:
    print(f"Error descargando {tick}: {e}")

df_all_indices = pd.DataFrame(df_indices)
```

```python
df_all_indices.to_excel('precios_'+indice+'_'+fecha_inicio+'_'+fecha_fin+'.xlsx')
```
